```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#加载数据
#从图中可以明显看出有异常点
dataset = pd.read_csv("data/studentscores.csv")
X = dataset.iloc[:,:1].values
Y = dataset.iloc[:,1].values
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 1/4, random_state = 0)
sns.scatterplot(data=dataset,x="Hours",y="Scores")
```

```python
#训练集使用简单线性回归模型来训练
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(X_train,Y_train)
```

```python
#预测结果
Y_pred = regressor.predict(X_test)
Y_pred
```

```python
#结果可视化
#训练集可视化
train_data = np.column_stack((X_train,Y_train.T))
train_data = pd.DataFrame(train_data)
train_data.columns = ["Hour","Scores"]
# train_data
sns.scatterplot(data=train_data,x="Hour",y="Scores")
```

```python
#训练结果可视化
plt.scatter(X_train , Y_train, color = 'red')
plt.plot(X_train , regressor.predict(X_train), color ='blue')
plt.show()
```

```python
#测试结果可视化
plt.scatter(X_test , Y_test, color = 'red')
plt.plot(X_test , regressor.predict(X_test), color ='blue')
plt.show()

```
